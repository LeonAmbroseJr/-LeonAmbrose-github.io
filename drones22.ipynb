## Understanding Yaw and Rotor Movements

Let's break down what **yaw** is and how the drone's rotors work together to control its movement.

### What is Yaw?

**Yaw** is one of the three primary axes of rotation for an aircraft, including drones. It refers to the **rotation of the drone around its vertical axis**. Imagine a line going straight up and down through the center of the drone. Yaw is the movement of the drone turning left or right around this imaginary line.

*   **Positive Yaw:** Turning the nose of the drone to the right (clockwise rotation).
*   **Negative Yaw:** Turning the nose of the drone to the left (counter-clockwise rotation).

Think of it like turning your head left or right while standing still – you are rotating around your own vertical axis.

### How Rotor Movements Control Yaw and Other Movements

Most recreational drones, including quadcopters like the Contixo TD1 and Bokigibi SLRC F11 SERIES, have four rotors. The speed and direction of rotation of these rotors are precisely controlled by the drone's flight controller to achieve different movements:

*   **Hovering:** To hover, all four rotors spin at the same speed. Two rotors spin clockwise, and two spin counter-clockwise. This is crucial to counteract the rotational forces (torque) created by the spinning propellers; otherwise, the drone would just spin in the air. The pairs of rotors spinning in opposite directions are typically positioned diagonally from each other.

*   **Ascending/Descending (Throttle):** To go up, all four rotors increase their speed equally, generating more lift. To go down, all four rotors decrease their speed equally, reducing lift.

*   **Moving Forward/Backward (Pitch):** To move forward, the rear rotors spin slightly faster than the front rotors. This causes the drone to tilt forward, and the thrust from the rotors pushes it in that direction. To move backward, the front rotors spin slightly faster than the rear rotors, tilting the drone backward.

*   **Moving Left/Right (Roll):** To move right, the left rotors spin slightly faster than the right rotors, causing the drone to tilt to the right and move in that direction. To move left, the right rotors spin slightly faster than the left rotors, tilting the drone to the left.

*   **Controlling Yaw:** This is where the opposing rotation of the rotors comes into play. To yaw left or right, the drone changes the *relative* speed of the pairs of rotors spinning in opposite directions.
    *   **To Yaw Right (Clockwise):** The flight controller slightly increases the speed of the counter-clockwise spinning rotors and/or slightly decreases the speed of the clockwise spinning rotors. This imbalance in rotational force causes the drone to rotate clockwise (yaw right).
    *   **To Yaw Left (Counter-Clockwise):** The flight controller slightly increases the speed of the clockwise spinning rotors and/or slightly decreases the speed of the counter-clockwise spinning rotors. This imbalance causes the drone to rotate counter-clockwise (yaw left).

Understanding how these rotor speed changes translate into pitch, roll, yaw, and throttle is fundamental to learning how to control a drone effectively. Practice with your drone's controls will help you feel how these adjustments affect its movement in the air.
